```python
from scipy.sparse import csr_matrix
md=csr_matrix(mm.values)
from sklearn.neighbors import NearestNeighbors
model_knn=NearestNeighbors(metric='cosine',algorithm='brute')
model_knn.fit(md)
```

```python
mm.shape
```

```python
md.shape
```

```python
query_index=np.random.choice(mm.shape[0])
print(query_index)
distances,indices=model_knn.kneighbors(mm.iloc[query_index,:].values.reshape(1,-1),n_neighbors=6)
```

```python
mm.head()
```

```python
for i in range(0,len(distances.flatten())):
    if i==0:
        print('recommandations for {0}:\n'.format(md[query_index],"title"))
    else:
        print('{0}:{1},with distance of {2}'.format(i,md.index[indices.flatten()[i]],distances.flatten()[i]))
```
